# Question 5: Matrix Value Modification

```python
import pandas as pd
def multiply_matrix(input_df):
    result_df = input_df.copy()
    for column in result_df.columns:
        result_df[column] = result_df[column].apply(lambda x: x * 0.75 if x > 20 else x * 1.25)
    result_df = result_df.round(1)

    return result_df
df = pd.DataFrame({'A': [15, 25, 10], 'B': [30, 18, 22]})
result_df = multiply_matrix(df)
print(result_df)
```
